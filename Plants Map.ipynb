```python
mango_requirements = {
    "Scientific Name": "Mangifera indica",
    "Desc Life form": "tree",
    "Eco Temperat. requir. Optimal Min": 24,  # °C
    "Eco Temperat. requir. Optimal Max": 30,
    "Eco Temperat. requir. Absolute Min": 8,
    "Eco Temperat. requir. Absolute Max": 48,
    "Eco Rainfall (annual) Optimal Min": 600,  # mm
    "Eco Rainfall (annual) Optimal Max": 1500,
    "Eco Soil PH Optimal Min": 5.5,
    "Eco Soil PH Optimal Max": 7.5,
    "Eco Soil drainage Optimal": "well (dry spells)",
    "Eco Climate zone": ["tropical wet & dry (Aw)", "subtropical dry summer (Cs)"]
}



def find_mango_locations(plant_reqs, climate_df):
    """Find locations matching mango requirements with relaxed rainfall due to irrigation"""
    mask = (
        (climate_df['T2M_MIN'] >= plant_reqs['Eco Temperat. requir. Optimal Min'] - 5) &  # Temp buffer
        (climate_df['T2M_MAX'] <= plant_reqs['Eco Temperat. requir. Optimal Max'] + 5) &
        (climate_df['TSOIL1'] >= plant_reqs['Eco Soil PH Optimal Min']) &
        (climate_df['ALLSKY_SFC_SW_DWN'] >= 10)  # Mango needs high sunlight
    )
    
    # Egypt-specific: Ignore natural rainfall (assume irrigation)
    return climate_df.loc[mask, ['City', 'LAT', 'LON', 'T2M_MIN', 'T2M_MAX', 'TSOIL1']]

# Find suitable locations
mango_locations = find_mango_locations(mango_requirements, df)
print(f"Suitable locations for {mango_requirements['Scientific Name']} in Egypt:")
print(mango_locations.to_string(index=False))
```

```python
def classify_light_intensity(row):
    if (row['ALLSKY_SFC_SW_DWN'] > 250 and
        row['CLOUD_AMT_DAY'] < 15 and
        row['ALLSKY_SFC_UV_INDEX'] > 8 and
        row['CLRSKY_DAYS'] > 25):
        return 'very bright'
    elif (row['ALLSKY_SFC_SW_DWN'] > 200 and
          row['CLOUD_AMT_DAY'] < 30 and
          row['ALLSKY_SFC_UV_INDEX'] > 6 and
          row['CLRSKY_DAYS'] > 20):
        return 'clear skies'
    elif (row['ALLSKY_SFC_SW_DWN'] > 150 and
          row['CLOUD_AMT_DAY'] < 60 and
          row['ALLSKY_SFC_UV_INDEX'] > 4 and
          row['CLRSKY_DAYS'] > 10):
        return 'cloudy skies'
    else:
        return 'light shade'

# Apply the classification to the DataFrame
df_small['Light_Intensity_Category'] = df_small.apply(classify_light_intensity, axis=1)
```

```python
# Define map boundaries
min_lon, max_lon = 24.5, 37
min_lat, max_lat = 22, 32

# Create Map
m = Map(
    center=[(min_lat + max_lat) / 2, (min_lon + max_lon) / 2],
    zoom=6,
    layout=Layout(width="600px", height="600px"),
    basemap=ipyleaflet.basemaps.OpenStreetMap.Mapnik,
    max_bounds=[[min_lat, min_lon], [max_lat, max_lon]],
    max_bounds_visible=True
)

# Create a container for the legend (we'll use HTML for the legend)
legend = widgets.HTML()
legend.layout.margin = "0px 20px 20px 20px"

def get_matching_locations(year, month):
    """Find locations where monthly aggregated temperature matches plant requirements"""
    try:
        # Ensure Date column is datetime
        df_large['Date'] = pd.to_datetime(df_large['Date'])

        # Filter to selected year and month
        monthly_data = df_large[(df_large['Date'].dt.year == year) & (df_large['Date'].dt.month == month)].copy()
        if monthly_data.empty:
            return pd.DataFrame()

        # Group by location and calculate monthly min/max
        grouped = monthly_data.groupby(['LAT', 'LON']).agg({
            'T2M_MIN': 'mean',
            'T2M_MAX': 'mean'
        }).reset_index()

        matching_locations = []

        # Check each plant's optimal temp range
        for _, plant in df_plants.iterrows():
            mask = (
                (grouped['T2M_MIN'] >= plant['TOptimal_Min']) & 
                (grouped['T2M_MAX'] <= plant['TOptimal_Max'])
            )
            matches = grouped[mask].copy()

            if not matches.empty:
                matches['plant_name'] = plant['plant_name']
                matches['month'] = month
                matches['year'] = year
                matches['T2M_MIN'] = grouped['T2M_MIN']
                matches['T2M_MAX'] = grouped['T2M_MAX']
                matching_locations.append(matches)

        if matching_locations:
            return pd.concat(matching_locations)
        return pd.DataFrame()

    except Exception as e:
        print(f"Error processing month: {e}")
        return pd.DataFrame()

def update_map(change):
    year = year_slider.value
    month = month_slider.value

    matching_df = get_matching_locations(year, month)

    # Clear previous layers (except basemap)
    for layer in m.layers[1:]:
        m.remove(layer)

    if not matching_df.empty:
        unique_plants = matching_df['plant_name'].unique()
        colors = {plant: f"#{np.random.randint(0, 0xFFFFFF):06x}" for plant in unique_plants}

        # Create legend
        legend_html = "<div style='background: white; padding: 10px; border: 1px solid gray;'><h4>Plant Types</h4>"
        for plant, color in colors.items():
            legend_html += f"<div><span style='display: inline-block; width: 20px; height: 20px; background: {color}; margin-right: 5px;'></span>{plant}</div>"
        legend_html += "</div>"
        legend.value = legend_html

        for plant_name, group in matching_df.groupby('plant_name'):
            features = []
            for _, row in group.iterrows():
                feature = {
                    'type': 'Feature',
                    'geometry': {
                        'type': 'Point',
                        'coordinates': [row['LON'], row['LAT']]
                    },
                    'properties': {
                        'plant': plant_name,
                        'max_temp': row['T2M_MAX'],
                        'min_temp': row['T2M_MIN'],
                        'month': row['month'],
                        'year': row['year']
                    }
                }
                features.append(feature)

            geojson = {
                'type': 'FeatureCollection',
                'features': features
            }

            layer = GeoJSON(
                data=geojson,
                style={
                    'color': colors[plant_name],
                    'opacity': 1,
                    'weight': 1.9,
                    'fillOpacity': 0.5
                },
                point_style={
                    'radius': 8,
                    'fill': True,
                    'color': colors[plant_name],
                    'fill_color': colors[plant_name]
                },
                name=plant_name,
                hover_style={'fillOpacity': 0.8},
                tooltip=widgets.HTML("<b>Plant:</b> {plant}<br>"
                                   "<b>Year:</b> {year}<br>"
                                   "<b>Month:</b> {month}<br>"
                                   "<b>Max Temp:</b> {max_temp}°C<br>"
                                   "<b>Min Temp:</b> {min_temp}°C")
            )
            m.add_layer(layer)
    else:
        legend.value = "<div style='background: white; padding: 10px; border: 1px solid gray;'>No matching locations found</div>"

# Create sliders
year_slider = widgets.IntSlider(
    value=2019, min=2019, max=2023, step=1,
    description="Year:",
    style={'description_width': 'initial'}
)

month_slider = widgets.IntSlider(
    value=1, min=1, max=12, step=1,
    description="Month:",
    style={'description_width': 'initial'}
)

# Add observers
year_slider.observe(update_map, names='value')
month_slider.observe(update_map, names='value')

# Add layer control
m.add_control(LayersControl())

# UI layout
slider_box = widgets.HBox([year_slider, month_slider])
slider_container = widgets.VBox([slider_box], layout=Layout(justify_content='center', width='100%'))
map_and_legend = widgets.HBox([m, legend])
ui = widgets.VBox([slider_container, map_and_legend])

display(ui)

# Initialize the map
update_map(None)
```
