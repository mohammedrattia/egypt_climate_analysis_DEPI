```python
import folium
import numpy as np
import pandas as pd
from folium.plugins import HeatMap
from folium import plugins
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.colors as mcolors
```

```python
df = pd.read_csv("data/fullWithLocations_large.csv")
df.head()
```

```python
df=df[["Date","City","GOV","LAT","LON","YEAR",'Month',"TSOIL1","TSOIL2","TSOIL3","TSOIL4","TSOIL5","TSOIL6"]]
df.head()
```

```python

# Group by LAT, LON, YEAR, WOY and aggregate
Monthly_df = df.groupby(["LAT","LON","YEAR",'Month']).agg(
    City=("City","first"),
    GOV=("GOV","first"),
    Date=("Date","first"),
    AVG_SOIL1=('TSOIL1', 'mean'),
    AVG_SOIL2=('TSOIL2', 'mean'),
    AVG_SOIL3=('TSOIL3', 'mean'),
    AVG_SOIL4=('TSOIL4', 'mean'),
    AVG_SOIL5=('TSOIL5', 'mean'),
    AVG_SOIL6=('TSOIL6', 'mean'),
).reset_index()

# Display the new DataFrame
Monthly_df=Monthly_df.sort_values(by=["YEAR",'Month',"LAT",'LON'],ascending=[True,True,True,True])
display(Monthly_df.head(10))
#Monthly_df.to_excel("TSOILS.xlsx",index=False)
```
