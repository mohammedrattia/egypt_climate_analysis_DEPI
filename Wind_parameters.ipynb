# Analysis with ipyleaflet

```python
# Define map boundaries
min_lon, max_lon = 24.5, 37
min_lat, max_lat = 22, 32

# Create Map
m = ipyleaflet.Map(
    center=[(min_lat + max_lat) / 2, (min_lon + max_lon) / 2],  # Centered within bounds
    zoom=6,
    layout=Layout(width="600px", height="600px"),  # Set fixed size
    basemap=ipyleaflet.basemaps.OpenStreetMap.Mapnik,
    max_bounds=[[min_lat, min_lon], [max_lat, max_lon]],  # Restrict panning
    max_bounds_visible=True  # Prevents moving outside max/min bounds
)

# Ensure df is a full copy to avoid warnings
df = df.copy()
df.loc[:, "ZWC2M"] = df["WS2M"] * np.cos(df["WD2M"])
df.loc[:, "MWC2M"] = df["WS2M"] * np.sin(df["WD2M"])

# Initialize global wind variable
wind = None  

# Function to update wind velocity layer
def update_wind(change):
    global wind  
    year = year_slider.value
    doy = day_slider.value  
    
    # Extract data for selected year & day
    wind_data = df[(df["YEAR"] == year) & (df["DOY"] == doy)].copy()  
    
    if wind_data.empty:
        return  
    
    wind_layer = Velocity(
        data=wind_data[["LAT", "LON", "ZWC2M", "MWC2M"]].set_index(["LAT", "LON"]).to_xarray(),
        zonal_speed="ZWC2M",
        meridional_speed="MWC2M",
        latitude_dimension="LAT",
        longitude_dimension="LON",
        velocity_scale=0.05,
        max_velocity=10
    )
    
    if wind is not None:
        m.remove(wind)
    m.add(wind_layer)
    
    wind = wind_layer  

# Create Year Slider (2019–2023)
year_slider = widgets.IntSlider(
    value=2019, min=2019, max=2023, step=1, description="Year"
)
year_slider.observe(update_wind, names="value")

# Create Day Slider (DOY: 1–365)
day_slider = widgets.IntSlider(
    value=1, min=1, max=365, step=1, description="Day"
)
day_slider.observe(update_wind, names="value")

# Display map and sliders
ui = widgets.VBox([year_slider, day_slider])
display(m, ui)

# Initialize the first wind layer
update_wind({"new": 1})
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from IPython.display import display, Markdown


def generate_visuals(df, attribute):
    for YEAR in df["YEAR"].unique():
        dff = df[df["YEAR"] == YEAR]  # Filter data for the specific year

        # Markdown for Histogram
        display(Markdown(f"# <span style='color:#8B4000; font-weight:bold;'>Histogram FOR {attribute} PER MONTH</span>"))
        
        # Histogram for attribute per month
        fig = px.histogram(dff, 
                           x=attribute, 
                           nbins=10,
                           facet_col="Month",
                           facet_col_wrap=5,
                           title=f"{attribute} Distribution for YEAR={YEAR}",  
                           labels={attribute: attribute, "count": "Frequency"},
                           color_discrete_sequence=["royalblue"],
                           opacity=0.75)
        fig.update_layout(template="plotly_white", xaxis_title=attribute, yaxis_title="Frequency", hovermode="x unified")
        fig.show()
        fig.write_html(f"resources/{attribute}/histogram_{attribute}_byMonth_Year{YEAR}.html")

        # Markdown for Pie Chart
        display(Markdown(f"# <span style='color:#8B4000; font-weight:bold;'>Pie Chart FOR {attribute} PER MONTH</span>"))
        
        # Pie Chart for Monthly Average
        df_avg = df.groupby(['YEAR', 'Month'], as_index=False)[attribute].mean()
        year_data = df_avg[df_avg['YEAR'] == YEAR]
        fig = px.pie(year_data, names='Month', values=attribute, 
                     title=f"Average {attribute} Distribution per Month in {YEAR}", hole=0.3)
        fig.update_traces(textinfo='label+percent')
        fig.show()
        fig.write_html(f"resources/{attribute}/piechart_{attribute}_byMonth_Year{YEAR}.html")

        # Markdown for Monthly Bar Graph
        display(Markdown(f"# <span style='color:#8B4000; font-weight:bold;'>BAR GRAPH FOR {attribute} PER MONTH</span>"))
        
        # Bar Graph for Monthly Min, Max, Mean
        all_months = pd.DataFrame({"Month": range(1, 13)})
        month_stats = dff.groupby("Month")[attribute].agg(["min", "max", "mean"]).reset_index()
        month_stats = all_months.merge(month_stats, on="Month", how="left").fillna(0)
        month_stats_melted = month_stats.melt(id_vars="Month", var_name="Statistic", value_name=attribute)
        fig = px.bar(month_stats_melted, x="Month", y=attribute, text=attribute, color="Statistic",
                     barmode="group", title=f"Min, Max, and Mean {attribute} per Month for {YEAR}",
                     color_discrete_map={"min": "blue", "max": "red", "mean": "green"})
        fig.show()
        fig.write_html(f"resources/{attribute}/months_MinMaxMean_Barplot_Year{YEAR}.html")

        # Markdown for City Bar Graph
        display(Markdown(f"# <span style='color:#8B4000; font-weight:bold;'>BAR GRAPH FOR {attribute} PER CITY</span>"))
        
        # Bar Graph for Cities Min, Max, Mean
        city_stats = dff.groupby("City")[attribute].agg(["min", "max", "mean"]).reset_index()
        city_stats_melted = city_stats.melt(id_vars="City", var_name="Statistic", value_name=attribute)
        fig = px.bar(city_stats_melted, x="City", y=attribute, color="Statistic", barmode="group",
                     title=f"Min, Max, and Mean {attribute} per City for {YEAR}",
                     color_discrete_map={"min": "#1f77b4", "max": "#ff7f0e", "mean": "#2ca02c"})
        fig.show()
        fig.write_html(f"resources/{attribute}/Cities_MinMaxMean_Barplot_Year{YEAR}.html")

    print(f"Visualizations generated for {attribute}!")

# Example usage
attributes = ["WD2M", "WD50M", "WS2M", "WS2M_MAX", "WS2M_MIN", "WS50M", "WS50M_MAX", "WS50M_MIN"]
for attr in attributes:
    generate_visuals(df, attr)
```
